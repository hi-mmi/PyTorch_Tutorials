```python
import torch
import numpy as np
```

```python
x = torch.tensor(1.0)
y = torch.tensor(2.0)

w = torch.tensor(1.0, requires_grad = True)

print(x,y,w)
```

```python
#forward pass
y_hat = x*w 
y_hat_loss = (y_hat - y)**2
print(y_hat_loss)
```

```python
#backwardpass
y_hat_loss.backward()
print(w.grad)
```
